```python
from util import parse_scenario_definition
import matplotlib.pyplot as plt
import geopandas as gpd

df = parse_scenario_definition()
```

```python

def plot_routes(df):
    fig, axes = plt.subplots(ncols=8, nrows=16, squeeze=True)
    fig.set_size_inches(16, 32)

    for (_, row), ax in zip(df.reset_index().iterrows(), axes.flatten()):
        p = row['points']
        ax.set_title(f"{row['route_index']}", fontsize=10)

        ax.set_aspect('equal')
        ax.set_xticks([])
        ax.set_yticks([])

        ax.plot(*p.xy, "o--", color='blue')

    # fig.savefig("plot.png", dpi=300)


plot_routes(df)
```

```python
df = gpd.GeoDataFrame(df, geometry='points')

for town, batch in df.groupby('town'):
    fig, ax = plt.subplots()
    # fig.set_size_inches(10, 10)
    ax.set_xticks([])
    ax.set_yticks([])
    batch.plot(ax=ax, cmap="tab10", linewidth=2)
```
